```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# config
GoldenRatio=(1+5**0.5)/2
# plt.rc('figure', figsize=(2, 2/GoldenRatio))
plt.rc('axes', grid=True)
# plt.rc('grid', color='0.1', linestyle=':', linewidth=0.3)
# matplotlib.rcParams.update({'font.size': 4})

plt.figure()    # open a new figure window

# line 1
plt.plot([2,3,4,8],[4,5,15,20], label="Line 1")

# line 2
x=np.linspace(0,10,100)
y=2*x+3
plt.plot(x,y,'-',label="Original line")

noise = np.random.randn(len(x))
noisy_flux = y + noise*5
plt.plot(x,noisy_flux,'k.')

p = np.polyfit(x,noisy_flux,1)
# plt.plot(x,p[0]*x+p[1],'r-',label="fit line")
plt.plot(x,np.polyval(p,x),'r-',label="Best fit line")

# other figure options 
plt.xlabel("X")
plt.ylabel("Y")
# plt.grid()
plt.legend(loc='best').get_frame().set_alpha(0.8)

# plt.savefig('/Volumes/Data/Xi/Downloads/aaaaa.pdf')
```
